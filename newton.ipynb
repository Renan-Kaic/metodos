```python
import numpy as np
from rich.console import Console
from rich.table import Table
from rich.panel import Panel
from rich.markdown import Markdown
from rich import box
import time
import matplotlib.pyplot as plt

# Configurar numpy para não usar notação científica
np.set_printoptions(precision=5, suppress=True)

# Inicializar o console Rich
console = Console()

def newton_method(f, df, x0, tol=1e-6, max_iter=100, show_steps=False):
    """
    Resolve uma equação não linear usando o método de Newton.
    
    Parâmetros:
    f: função para encontrar a raiz
    df: derivada da função
    x0: estimativa inicial
    tol: tolerância para convergência
    max_iter: número máximo de iterações
    show_steps: se True, mostra os passos intermediários do método
    
    Retorna:
    x: raiz encontrada
    iterations: número de iterações realizadas
    time_history: histórico de tempo de operações
    error_msg: mensagem de erro (se houver)
    x_history: histórico de valores de x
    """
    x = x0
    x_history = [x0]
    time_history = []
    start_time = time.time()
    
    if show_steps:
        console.print(f"[bold cyan]Iniciando o Método de Newton com x0 = {x0:.5f}[/bold cyan]")
    
    for i in range(max_iter):
        # Calcular o valor da função e sua derivada no ponto atual
        fx = f(x)
        dfx = df(x)
        
        # Verificar se a derivada é próxima de zero
        if abs(dfx) < 1e-10:
            error_msg = "Derivada próxima de zero. O método pode estar divergindo ou o ponto pode ser um ponto de inflexão."
            return x, i, time_history, error_msg, x_history
        
        # Calcular o próximo valor de x
        x_new = x - fx / dfx
        
        # Registrar o tempo após cada iteração
        current_time = time.time() - start_time
        time_history.append(current_time)
        
        # Adicionar o novo valor ao histórico
        x_history.append(x_new)
        
        if show_steps:
            console.print(f"[green]Iteração {i+1}: x = {x:.5f}, f(x) = {fx:.5f}, f'(x) = {dfx:.5f}, x_novo = {x_new:.5f}[/green]")
        
        # Verificar convergência
        if abs(x_new - x) < tol:
            return x_new, i+1, time_history, None, x_history
        
        # Atualizar x para a próxima iteração
        x = x_new
    
    error_msg = f"O método não convergiu após {max_iter} iterações."
    return x, max_iter, time_history, error_msg, x_history

def mostrar_resultados(f, df, x0, x, iterations, time_history, error_msg, x_history, tol, max_iter, nome_funcao):
    """
    Mostra os resultados do método de Newton.
    
    Parâmetros:
    f: função para encontrar a raiz
    df: derivada da função
    x0: estimativa inicial
    x: raiz encontrada
    iterations: número de iterações realizadas
    time_history: histórico de tempo de operações
    error_msg: mensagem de erro (se houver)
    x_history: histórico de valores de x
    tol: tolerância para convergência
    max_iter: número máximo de iterações
    nome_funcao: nome da função para exibição
    """
    # Mostrar resultados principais
    table = Table(title=f"Resultados do Método de Newton para {nome_funcao}")
    table.add_column("Parâmetro", style="cyan")
    table.add_column("Valor", style="green")
    
    table.add_row("Estimativa inicial (x0)", f"{x0:.5f}")
    table.add_row("Raiz encontrada", f"{x:.5f}")
    table.add_row("Valor da função na raiz", f"{f(x):.5f}")
    table.add_row("Iterações realizadas", str(iterations))
    table.add_row("Tempo total", f"{time_history[-1]:.5f} segundos")
    table.add_row("Tolerância", f"{tol:.5e}")
    table.add_row("Máximo de iterações", str(max_iter))
    
    console.print(table)
    
    # Mostrar histórico de iterações
    iter_table = Table(title="Histórico de Iterações")
    iter_table.add_column("Iteração", style="cyan")
    iter_table.add_column("x", style="green")
    iter_table.add_column("f(x)", style="blue")
    iter_table.add_column("f'(x)", style="magenta")
    iter_table.add_column("Erro", style="red")
    
    for i in range(len(x_history) - 1):
        x_atual = x_history[i]
        x_prox = x_history[i+1]
        erro = abs(x_prox - x_atual)
        iter_table.add_row(
            str(i+1),
            f"{x_atual:.5f}",
            f"{f(x_atual):.5f}",
            f"{df(x_atual):.5f}",
            f"{erro:.5e}"
        )
    
    console.print(iter_table)
    
    # Conclusão
    if error_msg is None:
        console.print(Panel(
            f"[bold green]✓ Método de Newton convergiu em {iterations} iterações![/bold green]\n"
            f"Raiz encontrada: {x:.5f}\n"
            f"Valor da função na raiz: {f(x):.5e}\n"
            f"Tempo total: {time_history[-1]:.5f} segundos",
            title="Conclusão",
            border_style="green"
        ))
    else:
        console.print(Panel(
            f"[bold yellow]⚠️ {error_msg}[/bold yellow]\n"
            f"Último valor calculado: {x:.5f}\n"
            f"Valor da função: {f(x):.5e}",
            title="Conclusão",
            border_style="yellow"
        ))
    
    # Sugestão para relatório
    console.print(Markdown("""
    ## Sugestão para anotação no relatório:
    
    ### Método Utilizado
    O problema foi resolvido utilizando o método iterativo de Newton para encontrar raízes de equações não lineares.
    
    ### Parâmetros da Solução
    - **Função analisada:** {}
    - **Estimativa inicial:** {:.5f}
    - **Tolerância:** {:.5e}
    - **Número de iterações:** {}
    - **Tempo de processamento:** {:.5f} s
    
    ### Solução Encontrada
    Raiz x = {:.5f}
    
    ### Verificação
    O valor da função na raiz encontrada é f(x) = {:.5e}
    
    ### Conclusão
    {}
    """.format(
        nome_funcao, 
        x0,
        tol,
        iterations,
        time_history[-1],
        x,
        f(x),
        "A raiz encontrada é exata e confiável." if error_msg is None else 
        "O método encontrou uma aproximação, mas com limitações: " + error_msg
    )))

def plotar_convergencia(x_history, f, nome_funcao):
    """
    Plota os gráficos de convergência do método de Newton.
    
    Parâmetros:
    x_history: histórico de valores de x
    f: função analisada
    nome_funcao: nome da função para exibição
    """
    try:
        # Preparar dados para plot
        iteracoes = range(len(x_history))
        erros = [abs(x_history[i+1] - x_history[i]) for i in range(len(x_history)-1)]
        erros.insert(0, 0)  # Adicionar um valor para a iteração 0
        
        # Criar subplots
        fig, (ax1, ax2, ax3) = plt.subplots(3, 1, figsize=(10, 15))
        
        # Gráfico de convergência de x
        ax1.plot(iteracoes, x_history, 'bo-')
        ax1.set_xlabel('Iteração')
        ax1.set_ylabel('Valor de x')
        ax1.set_title(f'Convergência do Método de Newton para {nome_funcao}')
        ax1.grid(True)
        
        # Gráfico de erro absoluto
        ax2.semilogy(iteracoes[1:], erros[1:], 'ro-')
        ax2.set_xlabel('Iteração')
        ax2.set_ylabel('Erro Absoluto (escala log)')
        ax2.set_title('Erro Absoluto por Iteração')
        ax2.grid(True)
        
        # Gráfico da função
        x_min = min(x_history) - 0.5
        x_max = max(x_history) + 0.5
        x_vals = np.linspace(x_min, x_max, 100)
        y_vals = [f(x) for x in x_vals]
        
        ax3.plot(x_vals, y_vals, 'g-')
        ax3.plot(x_history, [f(x) for x in x_history], 'ro')
        ax3.axhline(y=0, color='k', linestyle='-', alpha=0.3)
        ax3.set_xlabel('x')
        ax3.set_ylabel('f(x)')
        ax3.set_title(f'Função {nome_funcao} e Iterações do Método de Newton')
        ax3.grid(True)
        
        plt.tight_layout()
        plt.show()
        
    except Exception as e:
        console.print(f"[yellow]Não foi possível gerar os gráficos: {str(e)}[/yellow]")

def main():
    console.print(Markdown("# Método de Newton para Equações Não Lineares"), style="bold blue")
    
    try:
        # Definir funções e suas derivadas
        
        # Função 1: f(x) = x^3 - 2x - 5
        def f1(x):
            return x**3 - 2*x - 5
        
        def df1(x):
            return 3*x**2 - 2
        
        # Função 2: f(x) = cos(x) - x
        def f2(x):
            return np.cos(x) - x
        
        def df2(x):
            return -np.sin(x) - 1
        
        # Definir parâmetros para o método
        tol = 1e-6
        max_iter = 100
        mostrar_passos = True
        
        # Resolver primeira função
        console.rule("[bold cyan]Função 1: f(x) = x^3 - 2x - 5 [/bold cyan]")
        console.print("\n[bold cyan]Executando o Método de Newton para f(x) = x^3 - 2x - 5...[/bold cyan]")
        
        x0_1 = 2.0  # Estimativa inicial
        
        inicio = time.time()
        x1, iteracoes1, tempo_hist1, erro1, x_hist1 = newton_method(f1, df1, x0_1, tol, max_iter, mostrar_passos)
        fim = time.time()
        
        # Mostrar resultados
        mostrar_resultados(f1, df1, x0_1, x1, iteracoes1, tempo_hist1, erro1, x_hist1, tol, max_iter, "f(x) = x^3 - 2x - 5")
        
        # Plotar convergência
        try:
            plotar_convergencia(x_hist1, f1, "f(x) = x^3 - 2x - 5")
        except Exception as e:
            console.print(f"[yellow]Não foi possível gerar os gráficos: {str(e)}[/yellow]")
        
        # Resolver segunda função
        console.rule("[bold cyan]Função 2: f(x) = cos(x) - x [/bold cyan]")
        console.print("\n[bold cyan]Executando o Método de Newton para f(x) = cos(x) - x...[/bold cyan]")
        
        x0_2 = 0.5  # Estimativa inicial
        
        inicio = time.time()
        x2, iteracoes2, tempo_hist2, erro2, x_hist2 = newton_method(f2, df2, x0_2, tol, max_iter, mostrar_passos)
        fim = time.time()
        
        # Mostrar resultados
        mostrar_resultados(f2, df2, x0_2, x2, iteracoes2, tempo_hist2, erro2, x_hist2, tol, max_iter, "f(x) = cos(x) - x")
        
        # Plotar convergência
        try:
            plotar_convergencia(x_hist2, f2, "f(x) = cos(x) - x")
        except Exception as e:
            console.print(f"[yellow]Não foi possível gerar os gráficos: {str(e)}[/yellow]")
        
    except Exception as e:
        console.print(f"[bold red]Erro:[/bold red] {str(e)}")

if __name__ == "__main__":
    main()
```
